```python
import sys
sys.path.insert(0, 'ThinkDSP/code/') 
import thinkdsp
import matplotlib.pyplot as pyplot
import IPython
import thinkplot
```

```python
wave = thinkdsp.read_wave('tabla10.wav')
wave.plot(color='green')
thinkplot.config(xlabel='Time (S)', ylabel='Amplitude', ylim=[-1.25, 1.25])
```

```python
wave.make_audio()
```

```python
clipLength = 10 # in seconds
index = 0
while (index < wave.ts.size and wave.ts[index] < clipLength):
	index += 1
# Remove extras
wave.ts = wave.ts[:index]
wave.ys = wave.ys[:index]
```

```python
spectrum = wave.make_spectrum()
spectrum.low_pass(cutoff = 100, factor = .1)
#spectrum.high_pass(cutoff = 1500, factor = .1) # FIXME - Change back to low pass filter
filteredWave = spectrum.make_wave()
```

```python
spectrum = filteredWave.make_spectrum()
spectrum.plot(color='magenta')
thinkplot.config(xlabel='Frequency (Hz)', xlim=[-200, 20000], ylabel='Amplitude', ylim=[-200, 8000])
```

```python
spectrum.peaks()[:5]
```

```python
filteredWave.play()
IPython.display.Audio('sound.wav')
```
